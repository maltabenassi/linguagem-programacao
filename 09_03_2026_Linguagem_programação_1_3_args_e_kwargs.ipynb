```python
def calcular_media(*notas):
    # 'notas' chega como uma tupla: (10, 8, 9)
    total = sum(notas)
    quantidade = len(notas)
    media = total / quantidade
    return media

# Uso:
print(f"Média Semestre 1: {calcular_media(8.5, 9.0, 7.5)}")
print(f"Média Semestre 2: {calcular_media(10, 9.5)}")
```

A grosso modo, o *args serve para quando eu quero fazer uma lista com varios itens, porém sem querer contar quantos itens eu tenho dentro dessa lista, basicamente funcionando como um funil.

Exemplo: def parametros_aleatorios(*numeros):
    print(numeros)

# Chamando a função
parametros_aleatorios(1, 2, 3, 4, 5)


```python
def cadastrar_produto(nome, preco, **detalhes):
    print(f"Produto: {nome} | Preço: R${preco}")

    # Percorre as etiquetas (chaves) e os valores
    for caracteristica, valor in detalhes.items():
        print(f"- {caracteristica.capitalize()}: {valor}")

# Uso:
cadastrar_produto("Smartphone", 2500, cor="Grafite", memoria="128GB", rj45=False)
print("-" * 30)
cadastrar_produto("Monitor", 1200, polegadas=27, frequencia="144Hz")
```

O **kwargs, serve basicamente da mesma forma que o *args, porém como esses itens DEVEM ter etiquetas, pois se não a lista da erro.

Exemplo: def parametros_aleatorios(**numeros):
    print(numeros)

parametros_aleatorios(a=1, b=2, c=3, d=4, e=5)
# Saída: {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}  → Dicionário com pares chave-valor.
