```python
import random

numero_secreto = random.randint(1, 20)
tentativas = 0
max_tentativas = 5
vitoria = False

print("--- DESAFIO DO NÚMERO SECRETO ---")

while tentativas < max_tentativas:
    try:
        chute = int(input(f"\nTentativa {tentativas + 1}: Digite um número (1-20): "))
    except ValueError:
        print("Digite um número inteiro válido.")
        continue

    tentativas += 1

    if chute == numero_secreto:
        print(f"Exato! Você acertou em {tentativas} tentativas.")
        vitoria = True
        break
    elif chute < numero_secreto:
        print("Dica: O número secreto é MAIOR.")
    else:
        print("Dica: O número secreto é MENOR.")

if tentativas == max_tentativas and not vitoria:
    print(f"\nFim de jogo! O número era {numero_secreto}.")
```

O Objetivo desse "Jogo" é demonstrar na pratica alguns dos conceitos basicos de programação referentes ao capitulo 1.2 do livro que abordam estruturas "condicionais e de repetição".

Como por exemplo:
while = Utilizada para criar um ciclo de execução que permanece ativo enquanto uma condição lógica for verdadeira (neste caso, enquanto o usuário tiver tentativas disponíveis e não tiver acertado o número).

if,elif e else = Aplicadas para realizar a tomada de decisão do algoritmo. O programa compara o valor inserido pelo usuário com o valor armazenado, decidindo qual mensagem exibir (Maior, Menor ou Acerto).

Operadores Relacionais e Lógicos = Uso de operadores como == (igualdade), < (menor que) e o operador lógico and para verificar múltiplas condições simultâneas (como o fim das tentativas e o status de vitória).

Tipagem Dinâmica e Conversão: Demonstração da função input() para entrada de dados e a função int() para a conversão de string para inteiro, permitindo operações matemáticas e comparativas.
