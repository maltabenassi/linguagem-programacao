```python
import random

print("--- JOGO DE ADIVINHAÇÃO 2.0: EDIÇÃO LÓGICA ---")
print("Escolha sua dificuldade:")
print("1 - Fácil (10 tentativas)")
print("2 - Médio (5 tentativas)")
print("3 - Difícil (3 tentativas)")

# Aplicando a conversão de tipo (int) e input da Seção 1.2
opcao = int(input("\nDigite o número da opção: "))

# Estrutura Encadeada (elif) para definir o fluxo do jogo
if opcao == 1:
    max_tentativas = 10
    print("Modo Fácil ativado!")
elif opcao == 2:
    max_tentativas = 5
    print("Modo Médio ativado!")
elif opcao == 3:
    max_tentativas = 3
    print("Modo Difícil ativado!")
else:
    print("Opção inválida! Definindo padrão (Médio).")
    max_tentativas = 5

numero_secreto = random.randint(1, 20)
tentativas = 0
vitoria = False

# Estrutura de Repetição (while) conforme a Figura 1.7 do livro
while tentativas < max_tentativas:
    chute = int(input(f"\nTentativa {tentativas + 1} de {max_tentativas}: "))
    tentativas += 1

    if chute == numero_secreto:
        print(f"Parabéns! Você venceu com {tentativas} tentativas.")
        vitoria = True
        break
    elif chute < numero_secreto:
        print("Dica: O número secreto é MAIOR.")
    else:
        print("Dica: O número secreto é MENOR.")

# Operador Lógico (and) para verificar o fim do jogo
if tentativas == max_tentativas and not vitoria:
    print(f"\nGame Over! O número era {numero_secreto}.")
```
